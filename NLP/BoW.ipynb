```python
df.rename(columns={"v1": "label", "v2": "message"},inplace=True)
```

```python
import re
import nltk
nltk.download("stopwords")
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python

from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
```

```python
corpus = []
for i in range(len(df)):
    # Remove non-alphabetic characters, retain spaces
    sms = re.sub('[^a-zA-Z ]', '', df['message'][i])
    # Convert to lower case
    sms = sms.lower()
    # Tokenize the message
    words = nltk.word_tokenize(sms)
    # Remove stopwords, apply stemming and lemmatization
    words = [lemmatizer.lemmatize(stemmer.stem(word)) for word in words if word not in stop_words]
    # Join the words back into a single string
    sms = " ".join(words)
    # Append to the corpus
    corpus.append(sms)
```
